```python
import pandas as pd
#pandas is a library and it contains the functions related to the database
#importing the functions from the library
```

```python
dataset=pd.read_csv("salary_Data.csv")
#dataset is called and assigned to the variable
```

```python
dataset
#viewing the dataset
```

```python
#assigning the independant variable
independent=dataset[["YearsExperience"]]
#Viewing the variable
independent
```

```python
#assigning the dependant variable
dependent=dataset[["Salary"]]
#viewing the variable
dependent
```

```python
#spliting the test set and training set from the SKlearn library
#calling the imported function
#independent is input and dependent is output variable
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent, dependent, test_size=0.30, random_state=0)
```

```python
#importing the LinearRegression from the sklearn library
from sklearn.linear_model import LinearRegression
#assigning the non parameterized function for regressor
#LinearRegression is a class, all the functions under the LinearRegression will be assigned to regressor
regressor=LinearRegression()
#calling the fit function/method from the regressor by inserting the input (x_train) and output (y_train) for training set 
regressor.fit(X_train,y_train)
```

```python
#assigning the variable for calculating the weight. Weight is saved in the name of regressor.coef_
weight=regressor.coef_
#calling the variable
weight
```

```python
#assigning the variable for bais. Bais is saved in the name of regressor.intercept_
bais=regressor.intercept_
#calling the variable
bais
```

```python
#creating the variable. Here "predict" is the function for application and "x_test" is the input data.
y_pred=regressor.predict(X_test)
```

```python
#To find out the r_score value, importing the r2_score from the sklearn library by using the function
from sklearn.metrics import r2_score
#creating the variable for r_score using the actual (y_test) and predicted (y_pred) output
r_score=r2_score(y_test,y_pred)
```

```python
#To run the function
r_score
```

```python
#saving the model. import from pickle library
import pickle
#assigning the filename and saving it in the pickle format using ".sav" as extension
filename="finalized_model_linear.sav"
```

```python
#inserting the model to filename. All the models are saved under "regressor" so using this term to create the file. th term "wb" means write binary.
#model saved
pickle.dump(regressor,open(filename,'wb'))
```

```python
#to load the saved model
#rb means read binary
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
#assigning the variable for model prediction
result=loaded_model.predict([[13]])
```
